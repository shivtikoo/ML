# HOUSE PRICE PREDICTION 

- USING LINEAR REGRESSION  
        - SINGLE VARIABLE LINEAR REGRESSION


- PROBLEM STATEMENT :
    - PREDICTING THE PRICE OF THE HOUSE DEPENDENT ON THE AREA OF THE HOUSE !!!


- WE TRY TO VISUALIZE THE DATASET


- LINEAR REGRESSION 
    - WE TRY TO MODEL DATA USING A STRAIGHT LINE !!!
    - A STRAIGHT LINE WHICH COVERS THE MAXIMUM OF DATA POINTS
    - M = SUMMATION{(Y-YM)(X-XM)/(X-XM)^2}
    

```python
# IMPORTING THE REQUIRED LIBRARIES 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

dataset = pd.read_csv('dataset.csv')

print(dataset.shape)
print(dataset.head(5))

# SPLITTING THE DATASET ACCORDING TO OUR REQUIREMENT 

X = dataset.drop('price', axis = 'columns')
Y = dataset.iloc[:,1]

# DO A PLOT OF THE DATASET

plt.xlabel('AREA ')
plt.ylabel('PRICE ')
plt.scatter(X,Y,color='blue',marker='*')
```

```python
# TRAINING THE MODEL 

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X,Y)

# MAKING PREDICTION FOR A CUSTOM INPUT 

x=[[500]]
PredictedmodelResult = model.predict(x)

print(PredictedmodelResult)

```

```python
# USING THE FITTED LINE TO CHECK OUR PREDICTION

M = model.coef_
C = model.intercept_

y = M*x + C 

print('The predicted value using the best fit line is:')
print(y[0])

# CHECKING IF BOTH THE PREDICTED VALUES ARE THE SAME 

if PredictedmodelResult == y[0]:
    print("\nTHE MODEL WORKS ON THE LOGIC OF THE BEST FITTED LINE !!!")
```
