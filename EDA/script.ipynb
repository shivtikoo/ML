```python
# NOW COMING TO THE FINAL STEP OF DATA VISUALIZATION LET'S GET A BETTER UNDERSTANDING OF THE DATASETS IN PICTURE !!!

# USING THE CODE FROM THE REFERENCE PROVIDED 

def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):

    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

```python
# MORE VISUALIZATIONS

# Visualizing distribution of 'cons_12m' in client_data
plt.figure(figsize=(8, 6))
sns.histplot(client_df['cons_12m'], kde=True)
plt.title('Distribution of Electricity Consumption (past 12 months)')
plt.xlabel('Electricity Consumption')
plt.ylabel('Frequency')
plt.show()

# Visualizing distribution of 'price_off_peak_var' in price_data
plt.figure(figsize=(8, 6))
sns.histplot(price_df['price_off_peak_var'], kde=True)
plt.title('Distribution of Energy Price (off peak)')
plt.xlabel('Energy Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# NOW LET'S START WORKING TO THE MAIN GOAL THAT IS THE PRICE A FACTOR FOR THE CLIENT CHURNING !!!

# FILTERED DATAFRAMES FOR CHURNED AND NOT CHURNED CLIENTS !!!

churned = client_df[client_df['churn']==1]['id']
churned_clients = price_df[price_df['id'].isin(churned)]

not_churned = client_df[client_df['churn']==0]['id']
not_churned_clients = price_df[price_df['id'].isin(not_churned)]


# Plot for churned clients
plt.figure(figsize=(8, 6))
sns.histplot(churned_clients['price_off_peak_fix'], kde=True, color='red')
plt.title('Price of Power (First Period) - Churned Clients')
plt.xlabel('Price of Power')
plt.ylabel('Frequency')
plt.show()

# Plot for non-churned clients
plt.figure(figsize=(8, 6))
sns.histplot(not_churned_clients['price_off_peak_fix'], kde=True, color='green')
plt.title('Price of Power (First Period) - Non-Churned Clients')
plt.xlabel('Price of Power')
plt.ylabel('Frequency')
plt.show()
```

We are able to make out from the plot given the there is no direct correlation between the price of the power and the frequency of clients churning !!! 

```python
# LOOKING AT THE CORRELATION BETWEEN THE FEATURES AND THE CHURNING !!! 

# SEABORN 

numerical_data = client_df.select_dtypes(include='number').drop(columns='churn')

# Adding the 'Churn' column to the numerical data
numerical_data['churn'] = client_df['churn']
correlation_matrix = numerical_data.corr()

# HEATMAP FOCUSING ON CORRELATIONS WITH CHURN 
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix[['churn']], annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation with Churn')
plt.show()
```

```python
# TASK 03 VISUALIZATIONS !!!

# NOW LET US CROSS CHECK OUR ANALYSIS ABOUT NO STRONG CORRELATION EXISTING BETWEEN PRICE AND CHURN 

merged = price_df.merge(client_df[['id', 'churn']], on='id', how='inner')
merged = merged.drop(columns = ['id','price_date'])
correlation_matrix = merged.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix[['churn']], annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation with Churn')
plt.show()
```

SO THIS PLOT CROSS CHECKS OUR EARLIER ASSUMPTION ABOUT NO REAL CORRELATION BETWEEN CHURNIGN CLIENTS AND THE PRICE OF POWER !!!
