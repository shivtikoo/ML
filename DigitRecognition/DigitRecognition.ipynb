```python
# importing libraries
# for loading dataset
import pandas as pd
from sklearn.datasets import load_digits
```

```python
# loading the dataset
dataset = load_digits()
```

```python
# summarizing
print(dataset.data)
print(dataset.target)

print(dataset.data.shape)
print(dataset.images.shape)

dataimageLength = len(dataset.images)
print(dataimageLength)
```

**Visualising the Dataset**

```python
# choosing sample no. out of 1797
n=1796
import matplotlib.pyplot as plt
plt.gray()
plt.matshow(dataset.images[n])
plt.show()

dataset.images[n]
```

```python
# segregation
# input; pixel
X = dataset.images.reshape((dataimageLength,-1))
# output; class
Y = dataset.target
# not a csv file so no iloc
```

```python
# train n test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
print(X_train.shape)
print(X_test.shape)
```

**Training**

```python
from sklearn import svm
model = svm.SVC(kernel='linear')
model.fit(X_train,Y_train)
```

```python
# predicting
n = -2
# choosing numbers from behind as the first ones are for training
result = model.predict(dataset.images[n].reshape((1,-1)))
plt.imshow(dataset.images[n], cmap=plt.cm.gray_r, interpolation='nearest')
print(result)
print("/n")
plt.axis('off')
plt.title('%i' %result)
plt.show
```

```python
# prediction
Y_pred = model.predict(X_test)
import numpy as np
print(np.concatenate((Y_pred.reshape(len(Y_pred),1),Y_test.reshape(len(Y_test),1)),1))
```

```python
# evaluating
from sklearn.metrics import accuracy_score
score = accuracy_score(Y_pred,Y_test)*100
print('The accuracy of teh model is: ')
print(score)
```

```python
# experimenting
# using different svm kernels
model0 = svm.SVC(kernel='linear')
model1 = svm.SVC(kernel='rbf')
model2 = svm.SVC(gamma=0.001)
#model2 will also be rbf model just different gamma value as default gamma value is 0.1

model0.fit(X_train,Y_train)
model1.fit(X_train,Y_train)
model2.fit(X_train,Y_train)

Y_predmodel0 = model0.predict(X_test)
Y_predmodel1 = model1.predict(X_test)
Y_predmodel2 = model2.predict(X_test)

score0 = accuracy_score(Y_predmodel0,Y_test)*100
score1 = accuracy_score(Y_predmodel1,Y_test)*100
score2 = accuracy_score(Y_predmodel2,Y_test)*100

print('Accuracy of Model0 is: ')
print(score0)
print('Accuracy of Model1 is: ')
print(score1)
print('Accuracy of Model2 is: ')
print(score2)
```
