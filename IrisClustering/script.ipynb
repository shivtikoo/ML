# IRIS CLUSTERING 
    
    - PRINCIPAL COMPONENT ANALYSIS
         
        - IT WORKS ON THE IDEA OF DIMENSIONALITY REDUCTION; WHICH CONTAINS MOST OF THE INFORMATION OF THE DATASET 

        - STEPS FOR THE ALGORITHM:
            - FORM THE COVARIANCE MATRIX 
            - FIND THE EIGEN VECTORS & ORDERING THEM BY THEIR EIGEN VALUES IN DESCENDING ORDER !!!    

```python
# IMPORTING STANDARD LIBRARIES 

import pandas as pd
from sklearn import datasets 
import matplotlib.pyplot as plt

# LOADING THE DATASET 

data = datasets.load_iris()

X = data.data
y = data.target
names = data.target_names
```

```python
# PCA

from sklearn.decomposition import PCA

model = PCA(n_components = 3)
y_means = model.fit(X).transform(X)
```

```python
# VISUALISATION 

plt.figure()
colors = ['red', 'blue', 'yellow']

for color, i, target_name in zip(colors, [0, 1, 2], names):
    plt.scatter(y_means[y == i, 0], y_means[y == i, 1], color=color, lw = 2, label=target_name)

plt.title('IRIS CLUSTERS ')
plt.show()
```
