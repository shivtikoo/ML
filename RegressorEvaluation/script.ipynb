# REGRESSION MODEL EVALUATION METRICS

     - R-squared (RÂ²):

         - What it tells us: 
         R-squared measures how well the regression model fits the observed data.

         - Easy Explanation: 
         Imagine you're trying to predict students' test scores based on the number of hours they study. R-squared tells you how much of the variability in the test scores can be explained by the hours of study.

         - Interpretation: 
         An R-squared value closer to 1 means that a larger proportion of the variation in the test scores can be explained by the hours of study. If it's closer to 0, the model doesn't explain much of the variability.

     - Adjusted R-squared:

         - What it tells us: 
         Adjusted R-squared is similar to R-squared but adjusts for the number of predictors in the model.

         - Easy Explanation: 
         Let's say you add more variables to predict the test scores, like sleep hours or extracurricular activities. Adjusted R-squared helps to account for whether these additional variables improve the model or just add noise.
        
         - Interpretation: 
         Higher values of adjusted R-squared indicate that the added predictors improve the model's fit, whereas lower values suggest the added variables might not be helpful.

     -  WORKS ON THE FORMULA 1 - SSR/SST
          - SSR : SUM OF SQUARED RESIDUALS DIFFERENCE BETWEEN Yi AND Ypred 
          - SST : SUM OF SQUARED TOTALS DIFFERENCE BETWEEN Yi AND Ymean 
         
In essence, R-squared tells you how well your model explains the variation in the target variable, while adjusted R-squared helps in assessing the model's complexity and whether adding more features benefits the model's performance.


```python
# IMPORTING THE STANDARD LIBRARIES

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
# READING THE TWO COMPARATIVE DATASETS 

data1 = pd.read_csv('dataset1.csv')
data2 = pd.read_csv('dataset2.csv')

print(data1.shape)
print(data1.head(5))

print(data2.shape)
print(data2.head(5))
```

```python
# TRAINING AND EVALUATING ON DATA1 

# SPLITTING THE DATASET 

X = data1.iloc[:,:-1]
Y = data1.iloc[:,-1]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)

# TRAINING THE DATASET 

model = LinearRegression()
model.fit(X_train,Y_train)

Y_pred =  model.predict(X_test)

print('R2 SCORE !!!')
x = r2_score(Y_test, Y_pred)
print(x)

print('ADJUSTED R2 SCORE !!!')
n=len(data1) #Length of Total dataset
p=len(data1.columns)-1 #length of Features
adjr= 1-(1-x)*(n-1)/(n-p-1)
print(adjr)

# PLOT THE GRAPH OF THE DATASET AND THE PREDICTIONS TO VISUALISE THE METRIC SCORES !!!

plt.scatter(X, Y, color = 'red')
plt.plot(X , model.predict(X))

plt.title('BEST LINE FOR THE DATASET ')
plt.xlabel('AREA')
plt.ylabel('PRICE')

plt.show()
```
