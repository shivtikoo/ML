# POLYNOMIAL REGRESSION
 
- FIT A POLYNOMIAL LINE TO THE DATASET 

```python
# IMPORT STANDARD LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# READ THE DATA

data = pd.read_csv('dataset.csv')

print(data.shape)
print(data.head(5))

# SPLITTING THE DATASET 

X = data.iloc[:,:-1].values
Y = data.iloc[:,-1].values
```

```python
# TRAINING THE LINEAR REGRESSOR
 
modelX = LinearRegression()
modelX.fit(X,Y)

# MAKING THE PLOT

plt.scatter(X,Y, color='red') 
plt.plot(X,(modelX.predict(X)))
plt.title('LINEAR REGRESSOR')
plt.xlabel('LEVEL')
plt.ylabel('SALARY')
plt.show()
```

```python
# TRAINING A POLYNOMIAL REGRESSOR

# TRANSFORMING THE TRAINING X

from sklearn.preprocessing import PolynomialFeatures

X_new = PolynomialFeatures(degree=3).fit_transform(X)

# CHECKING WHAT CHANGE THE TRAINING DATASET SEES !!
print(X_new)

ModelY = LinearRegression()
ModelY.fit(X_new,Y)

# PLOTTING THE POLYNOMIAL REGRESSOR !!!

plt.scatter(X,Y, color='yellow')
plt.plot(X,ModelY.predict(X_new))
plt.title('POLYNOMIAL REGRESSOR')
plt.xlabel('LEVEL')
plt.ylabel("SALARY")
plt.show()
```

- WE SEE HIGHER THE DEGREE OF THE CURVE IT THE BETTER THE FIT TO TEH DATASET !!!
